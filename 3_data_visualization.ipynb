```python
# Librerie
import os
import pandas as pd
import numpy as np
import folium
import matplotlib.pyplot as plt
plt.style.use('ggplot')
get_ipython().magic('pylab inline')
```

```python
# Cartelle Input/Output
dir_df = os.path.join(os.path.abspath(''),'stg')
dir_out = os.path.join(os.path.abspath(''),'output')
```

```python
df_g4_filename = r'df_morte.pkl'
df_g4_fullpath = os.path.join(dir_df, df_g4_filename)
df_g4 = pd.read_pickle(df_g4_fullpath)
```

```python
# Setto indice la Eta
df_g4 = df_g4.set_index(['Eta'])
```

```python
df_g4.head(5)
```

```python
tp = df_g4.plot(kind='line',
               logy = True,
               legend = False)
tp.set_xlabel("")
tp.set_ylabel("Numero di morti")
plt.xticks(rotation=45)
fig_posts = tp.get_figure()
fig_posts.tight_layout()
fig_posts.savefig(os.path.join(dir_out,'g4.png'), format='png', dpi=1000)
```
